```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x=np.arange(1,6)
y=np.array([3,4,2,4,5])
xmean=np.mean(x)
ymean=np.mean(y)
xar=np.array([])
yar=np.array([])
mul=np.array([])
sqr=np.array([])
lreg=np.array([])

for i in range (0,len(x)):
    xar=np.append(xar,x[i]-xmean)
    yar=np.append(yar,y[i]-ymean)
    mul=np.append(mul,xar[i]*yar[i])
    sqr=np.append(sqr,xar[i]**2)
# slope m
m=np.sum(mul)/np.sum(sqr)
print("m :> ",m)

#c y=mx+c
c=ymean-(m*xmean)
print("c :> ",c)

for i in range(0,len(x)):
   lreg= np.append(lreg,(m*x[i])+c)

```

```python
yp=np.array([])
ysq1=np.array([])
for i in range(0,len(x)):
    yp=np.append(yp,(lreg[i]-ymean)**2)
    ysq1=np.append(ysq1,(y[i]-ymean)**2)
r2=1-(np.sum(yp)/np.sum(ysq1))
print("R2 :> ",r2)
```

```python
plt.plot(x,y,'og')
plt.plot(x,lreg,linestyle="--")
```
