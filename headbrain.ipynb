```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv("C:\SemVI\ML\headbrain.csv")
data.head()
```

```python
hd=np.array(data["Head Size(cm^3)"])
br=np.array(data["Brain Weight(grams)"])
```

```python
hdmean=np.mean(hd)
brmean=np.mean(br)
brar=np.array([])
yar=np.array([])
mul=np.array([])
sqr=np.array([])
lreg=np.array([])

for i in range (0,len(hd)):
    brar=np.append(brar,hd[i]-hdmean)
    yar=np.append(yar,br[i]-brmean)
    mul=np.append(mul,brar[i]*yar[i])
    sqr=np.append(sqr,brar[i]**2)
# slope m
m=np.sum(mul)/np.sum(sqr)
print("m :> ",m)

#c br=mhd+c
c=brmean-(m*hdmean)
print("c :> ",c)

for i in range(0,len(hd)):
   lreg= np.append(lreg,(m*hd[i])+c)
```

```python
yp=np.array([])
ysq1=np.array([])
for i in range(0,len(hd)):
    yp=np.append(yp,(lreg[i]-brmean)**2)
    ysq1=np.append(ysq1,(br[i]-brmean)**2)
r2=1-(np.sum(yp)/np.sum(ysq1))
print("R2 :> ",r2)
```

```python
plt.plot(hd,br,'og')
plt.plot(hd,lreg,linestyle="--")
```
