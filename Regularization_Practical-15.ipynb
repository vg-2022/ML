```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
np.random.seed(40)
X = 2*np.random.rand(100,1)
y = 4+3*X+np.random.randn(100,1)
```

```python
plt.scatter(X,y,color='black')
plt.xlabel('X')
plt.ylabel('y')
plt.show()
```

```python
X_b = np.c_[np.ones((100,1)),X]
```

```python
theta1 = np.random.randn(2,1)
theta2 = np.random.randn(2,1)
learning_rate = 0.1
n_iterations = 1000
alpha = 0.1
```

```python
#Without-Regularization

loss_history=[]
for iteration in range(n_iterations):
    gradients = 2/100* X_b.T.dot(X_b.dot(theta1)-y)
    theta1 = theta1-learning_rate*gradients
    
    loss=np.mean((X_b.dot(theta1)-y)**2)
    loss_history.append(loss)
    
    
print("Final Theta :  ", theta1)

# Regularization

for iteration in range(n_iterations):
    gradients = 2/100 * X_b.T.dot(X_b.dot(theta2) - y) + 2 * alpha * theta2
    theta2 = theta2 - learning_rate * gradients
    loss = np.mean((X_b.dot(theta2) - y) ** 2) + alpha * np.sum(theta2[1:] ** 2)  # Regularized loss
    loss_history.append(loss)

print("Final Theta (Regularization): ", theta2)
```

```python
plt.scatter(X,y)
plt.plot(X,X_b.dot(theta1),label='Linear Regression without Regularization',color='red')
plt.plot(X,X_b.dot(theta2),label='Linear Regression with Regularization',color='green')
plt.xlabel('X')
plt.ylabel('y')
plt.legend()
plt.show()
```

```python
plt.plot(loss_history)
plt.ylabel("Loss")
plt.show()
```
